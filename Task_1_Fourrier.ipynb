## Spectral Analysis

```python
#Load the wav file
wf = wave.open('string_1.wav')
signal = np.fromstring(wf.readframes(-1), 'Int16')
#Calculate its spectrum with Fast Fourier Transform algorithm
freqs = fft(signal)

#Ploth both the wave and its spectrum
fig = plt.figure(figsize=(8, 6))
gs1 = gridspec.GridSpec(2, 1, height_ratios=[1, 1]) 
ax1, ax2 = fig.add_subplot(gs1[0]), fig.add_subplot(gs1[1])

ax1.plot(signal, color="#6480e5")
ax2.plot(np.abs(freqs), color="#6480e5")

ax1.set_title('Signal Wave'), ax2.set_title("Frequency Plot")
ax1.set_xlim([0, 1.5e5]), ax2.set_xlim([0, 6e3])
ax1.set_ylim([-0.5e4, 0.5e4]), ax2.set_ylim([0, 3e7])
ax1.set_xlabel('Sampling Points'), ax2.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Amplitude'), ax2.set_ylabel('Intensity') 
ax1.ticklabel_format(style='sci', axis='both', scilimits=(0,0)) 
ax2.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

plt.tight_layout()
```

## Automatic Note Detection

```python
def get_frequency(filename):
  sample_rate, data = scipy_wav.read(filename)
  N = len(data)
  n = 1
  energy = rfft(data)
  freqs = rfftfreq(N, 1 / sample_rate)
  freq = []
  idx = (-energy).argsort()[:n]
  for i in range(n) :
    freq.append(freqs[idx[i]])
  return np.min(freq)
  #print(freq)
  
```

```python
notes = {"E1" : 329.63, "B2" : 246.94 , "G3": 196.00, "D4": 146.83, "A5" : 110.00, "E6" : 82.41}
acc = {"E1" : 0.15, "B2" : 0.15 , "G3": 0.15, "D4": 0.08, "A5" : 0.08, "E6" : 0.04}
def NoteDetection(filename):
  freq = get_frequency(filename)
  temp = []
  for key,value in notes.items() :
    temp.append(np.abs(value-freq))
  note = [key for key,value in notes.items() if min(temp)+freq ==value or freq - min(temp)==value]
  accuracy = abs(notes[note[0]] -freq) /acc[note[0]]
  print( "The note is",note[0], "with an accuracy of" ,accuracy,"cents")
```

```python
NoteDetection('string_1.wav')
```
